```python
package com.example.calculatorapp;

import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;

import androidx.appcompat.app.AppCompatActivity;

public class InterestCalculatorActivity extends AppCompatActivity {
    EditText principal, years, rate, result;
    Button simpleInterest, compoundInterest;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_interest_calculator);

        principal = findViewById(R.id.principal);
        years = findViewById(R.id.years);
        rate = findViewById(R.id.rate);
        result = findViewById(R.id.result);
        simpleInterest = findViewById(R.id.simpleInterest);
        compoundInterest = findViewById(R.id.compoundInterest);

        simpleInterest.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                double p = Double.parseDouble(principal.getText().toString());
                double t = Double.parseDouble(years.getText().toString());
                double r = Double.parseDouble(rate.getText().toString());
                double si = (p * t * r) / 100;
                result.setText("Total Amount: " + (p + si));
            }
        });

        compoundInterest.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                double p = Double.parseDouble(principal.getText().toString());
                double t = Double.parseDouble(years.getText().toString());
                double r = Double.parseDouble(rate.getText().toString());
                double ci = p * Math.pow((1 + r / 100), t) - p;
                result.setText("Total Amount: " + (p + ci));
            }
        });
    }
}
```

```python
package com.example.calculatorapp;

import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;

import androidx.appcompat.app.AppCompatActivity;

import java.util.ArrayList;

public class PrimeSumActivity extends AppCompatActivity {
    EditText startNumber, range, primeNumbersWithSum;
    Button submit;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_prime_sum);

        startNumber = findViewById(R.id.startNumber);
        range = findViewById(R.id.range);
        primeNumbersWithSum = findViewById(R.id.primeNumbersWithSum);
        submit = findViewById(R.id.submit);

        submit.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                int start = Integer.parseInt(startNumber.getText().toString());
                int rangeValue = Integer.parseInt(range.getText().toString());
                int end = start + rangeValue;

                ArrayList<String> primesWithSum = getPrimesWithSumInRange(start, end);
                StringBuilder result = new StringBuilder();

                for (String entry : primesWithSum) {
                    result.append(entry).append("\n");
                }

                primeNumbersWithSum.setText(result.toString().trim());
            }
        });
    }

    private ArrayList<String> getPrimesWithSumInRange(int start, int end) {
        ArrayList<String> result = new ArrayList<>();
        for (int i = start; i <= end; i++) {
            if (isPrime(i)) {
                int digitSum = calculateSumOfDigits(i);
                result.add("Prime: " + i + " | Sum of Digits: " + digitSum);
            }
        }
        return result;
    }

    private boolean isPrime(int n) {
        if (n <= 1) return false;
        for (int i = 2; i <= Math.sqrt(n); i++) {
            if (n % i == 0) return false;
        }
        return true;
    }

    private int calculateSumOfDigits(int number) {
        int sum = 0;
        while (number > 0) {
            sum += number % 10;
            number /= 10;
        }
        return sum;
    }
}
```
