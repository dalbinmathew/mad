```python
package com.example.bankapp;

import android.content.Intent;
import android.content.SharedPreferences;
import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;


import com.example.bankapp.R;

public class MainActivity extends AppCompatActivity {
    EditText username, password;
    Button login, register;
    SharedPreferences sharedPreferences;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);

        username = findViewById(R.id.username);
        password = findViewById(R.id.password);
        login = findViewById(R.id.login);
        register = findViewById(R.id.register);

        sharedPreferences = getSharedPreferences("BankApp", MODE_PRIVATE);

        login.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                String storedUsername = sharedPreferences.getString("username", "");
                String storedPassword = sharedPreferences.getString("password", "");

                if (username.getText().toString().equals(storedUsername) &&
                        password.getText().toString().equals(storedPassword)) {
                    Intent intent = new Intent(MainActivity.this,com.example.bankapp.AccountActivity.class);
                    startActivity(intent);
                } else {
                    Toast.makeText(MainActivity.this, "Invalid username or password", Toast.LENGTH_SHORT).show();
                }
            }
        });

        register.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                Intent intent = new Intent(MainActivity.this, com.example.bankapp.RegistrationActivity.class);
                startActivity(intent);
            }
        });
    }
}
```

```python
package com.example.bankapp;

import android.content.Intent;
import android.content.SharedPreferences;
import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;
import android.widget.RadioButton;
import android.widget.RadioGroup;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;

import com.example.bankapp.R;

public class RegistrationActivity extends AppCompatActivity {
    EditText accountNumber, name, initialBalance, username, password;
    RadioGroup accountTypeGroup;
    RadioButton savings, current;
    Button registerButton;
    SharedPreferences sharedPreferences;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_registration);

        accountNumber = findViewById(R.id.accountNumber);
        name = findViewById(R.id.name);
        initialBalance = findViewById(R.id.initialBalance);
        username = findViewById(R.id.username);
        password = findViewById(R.id.password);
        accountTypeGroup = findViewById(R.id.accountTypeGroup);
        savings = findViewById(R.id.savings);
        current = findViewById(R.id.current);
        registerButton = findViewById(R.id.registerButton);

        sharedPreferences = getSharedPreferences("BankApp", MODE_PRIVATE);

        registerButton.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                if (validateFields()) {
                    SharedPreferences.Editor editor = sharedPreferences.edit();
                    editor.putString("accountNumber", accountNumber.getText().toString());
                    editor.putString("accountType", savings.isChecked() ? "Savings" : "Current");
                    editor.putString("name", name.getText().toString());
                    editor.putString("initialBalance", initialBalance.getText().toString());
                    editor.putString("username", username.getText().toString());
                    editor.putString("password", password.getText().toString());
                    editor.apply();

                    Toast.makeText(RegistrationActivity.this, "Registration Successful", Toast.LENGTH_SHORT).show();
                    startActivity(new Intent(RegistrationActivity.this, com.example.bankapp.MainActivity.class));
                }
            }
        });
    }

    private boolean validateFields() {
        if (accountNumber.getText().toString().isEmpty() ||
                name.getText().toString().isEmpty() ||
                initialBalance.getText().toString().isEmpty() ||
                username.getText().toString().isEmpty() ||
                password.getText().toString().isEmpty() ||
                accountTypeGroup.getCheckedRadioButtonId() == -1) {
            Toast.makeText(this, "All fields are required", Toast.LENGTH_SHORT).show();
            return false;
        }
        return true;
    }
}
```

```python
package com.example.bankapp;

import android.content.SharedPreferences;
import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;
import android.widget.TextView;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;

import com.example.bankapp.R;

public class AccountActivity extends AppCompatActivity {
    TextView accountDetails, balance;
    EditText amount;
    Button deposit, withdraw;
    SharedPreferences sharedPreferences;
    double currentBalance;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_account);

        accountDetails = findViewById(R.id.accountDetails);
        balance = findViewById(R.id.balance);
        amount = findViewById(R.id.amount);
        deposit = findViewById(R.id.deposit);
        withdraw = findViewById(R.id.withdraw);

        sharedPreferences = getSharedPreferences("BankApp", MODE_PRIVATE);
        currentBalance = Double.parseDouble(sharedPreferences.getString("initialBalance", "0"));

        String details = "Name: " + sharedPreferences.getString("name", "") + "\n" +
                "Account Type: " + sharedPreferences.getString("accountType", "") + "\n" +
                "Account Number: " + sharedPreferences.getString("accountNumber", "");
        accountDetails.setText(details);
        balance.setText("Balance: $" + currentBalance);

        deposit.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                double depositAmount = Double.parseDouble(amount.getText().toString());
                currentBalance += depositAmount;
                balance.setText("Balance: $" + currentBalance);
                Toast.makeText(AccountActivity.this, "Deposited: $" + depositAmount, Toast.LENGTH_SHORT).show();
            }
        });

        withdraw.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View v) {
                double withdrawAmount = Double.parseDouble(amount.getText().toString());
                if (withdrawAmount > currentBalance) {
                    Toast.makeText(AccountActivity.this, "Insufficient Balance", Toast.LENGTH_SHORT).show();
                } else {
                    currentBalance -= withdrawAmount;
                    balance.setText("Balance: $" + currentBalance);
                    Toast.makeText(AccountActivity.this, "Withdrew: $" + withdrawAmount, Toast.LENGTH_SHORT).show();
                }
            }
        });
    }
}
```
