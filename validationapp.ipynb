```python
package com.example.formvalidation;

import android.content.Intent;
import android.net.Uri;
import android.os.Bundle;
import android.view.View;
import android.widget.ArrayAdapter;
import android.widget.Button;
import android.widget.CheckBox;
import android.widget.EditText;
import android.widget.RadioButton;
import android.widget.RadioGroup;
import android.widget.Spinner;
import android.widget.TextView;
import android.widget.Toast;

import androidx.activity.EdgeToEdge;
import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {
    TextView nametv, emailtv, agetv, gendertv, yeartv, eventtv, phonetv, passwordtv;
    EditText name, email, age, phone, password;
    RadioGroup gendergroup;
    RadioButton male, female, other;
    CheckBox badminton, cricket, football;
    Spinner year;
    Button register, contact;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        EdgeToEdge.enable(this);
        setContentView(R.layout.activity_main);

        // Binding UI elements
        nametv = findViewById(R.id.nametv);
        emailtv = findViewById(R.id.emailtv);
        agetv = findViewById(R.id.agetv);
        gendertv = findViewById(R.id.gendertv);
        yeartv = findViewById(R.id.yeartv);
        eventtv = findViewById(R.id.eventtv);
        phonetv = findViewById(R.id.phonetv);
        passwordtv = findViewById(R.id.passwordtv);
        name = findViewById(R.id.name);
        email = findViewById(R.id.email);
        age = findViewById(R.id.age);
        phone = findViewById(R.id.phone);
        password = findViewById(R.id.password);
        gendergroup = findViewById(R.id.gendergroup);
        male = findViewById(R.id.male);
        female = findViewById(R.id.female);
        other = findViewById(R.id.other);
        badminton = findViewById(R.id.badminton);
        cricket = findViewById(R.id.cricket);
        football = findViewById(R.id.football);
        year = findViewById(R.id.year);
        register = findViewById(R.id.register);
        contact = findViewById(R.id.contact);

        // Spinner setup
        String[] years = new String[]{"1", "2", "3", "4"};
        ArrayAdapter<String> adapter = new ArrayAdapter<>(this, android.R.layout.simple_spinner_dropdown_item, years);
        year.setAdapter(adapter);

        // Register button click listener
        register.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View view) {
                String nums = "0123456789";
                String alphabets = "abcdefghijklmnopqrstuvwxyz";
                String special = "!@#$%^&*~|:',.></?";

                String name1 = name.getText().toString();
                String email1 = email.getText().toString();
                String age1 = age.getText().toString();
                String phone1 = phone.getText().toString();
                String password1 = password.getText().toString();


                if (name1.isEmpty()) {
                    Toast.makeText(MainActivity.this, "Name cannot be empty", Toast.LENGTH_SHORT).show();
                } else if (name1.matches(".*["+nums+"].*")) {
                    Toast.makeText(MainActivity.this, "Name cannot contain numbers", Toast.LENGTH_SHORT).show();
                } else if (!email1.contains("@")) {
                    Toast.makeText(MainActivity.this, "Email should contain @", Toast.LENGTH_SHORT).show();
                } else if (email1.isEmpty()) {
                    Toast.makeText(MainActivity.this, "Email cannot be empty", Toast.LENGTH_SHORT).show();
                } else if (age1.isEmpty()) {
                    Toast.makeText(MainActivity.this, "Age cannot be empty", Toast.LENGTH_SHORT).show();
                } else if (age1.matches(".*[a-zA-Z].*")) {
                    Toast.makeText(MainActivity.this, "Age cannot contain alphabets", Toast.LENGTH_SHORT).show();
                } else if (phone1.isEmpty()) {
                    Toast.makeText(MainActivity.this, "Phone number cannot be empty", Toast.LENGTH_SHORT).show();
                } else if (phone1.matches(".*[a-zA-Z].*")) {
                    Toast.makeText(MainActivity.this, "Phone number cannot contain alphabets", Toast.LENGTH_SHORT).show();
                } else if (phone1.length() != 10) {
                    Toast.makeText(MainActivity.this, "Phone number must be 10 digits", Toast.LENGTH_SHORT).show();
                } else if (password1.isEmpty()) {
                    Toast.makeText(MainActivity.this, "Password cannot be empty", Toast.LENGTH_SHORT).show();
                } else if (!password1.matches(".*[a-zA-Z].*")) {
                    Toast.makeText(MainActivity.this, "Password must contain at least one alphabet", Toast.LENGTH_SHORT).show();
                } else if (!password1.matches(".*\\d.*")) {
                    Toast.makeText(MainActivity.this, "Password must contain at least one number", Toast.LENGTH_SHORT).show();
                } else if (!password1.matches(".*[" + special + "].*")) {
                    Toast.makeText(MainActivity.this, "Password must contain at least one special character", Toast.LENGTH_SHORT).show();
                } else if (password1.length() < 8) {
                    Toast.makeText(MainActivity.this, "Password must be at least 8 characters", Toast.LENGTH_SHORT).show();
                } else if (gendergroup.getCheckedRadioButtonId() == -1) {
                    Toast.makeText(MainActivity.this, "Please select a gender", Toast.LENGTH_SHORT).show();
                } else if (!badminton.isChecked() && !cricket.isChecked() && !football.isChecked()) {
                    Toast.makeText(MainActivity.this, "Please select at least one event", Toast.LENGTH_SHORT).show();
                } else {
                    String year1 = year.getSelectedItem().toString();
                    String gender1 = "";
                    if (gendergroup.getCheckedRadioButtonId() != -1) {
                        RadioButton selectedGender = findViewById(gendergroup.getCheckedRadioButtonId());
                        gender1 = selectedGender.getText().toString();
                    }
                    StringBuilder events = new StringBuilder();
                    if (badminton.isChecked())
                        events.append("Badminton ");
                    if (cricket.isChecked())
                        events.append("Cricket ");
                    if (football.isChecked())
                        events.append("Football");

                    Intent intent = new Intent(MainActivity.this, SecondActivity.class);
                    intent.putExtra("name", name1);
                    intent.putExtra("email", email1);
                    intent.putExtra("age", age1);
                    intent.putExtra("phone", phone1);
                    intent.putExtra("password", password1);
                    intent.putExtra("year", year1);
                    intent.putExtra("gender", gender1);
                    intent.putExtra("events", events.toString());
                    startActivity(intent);
                }

            }});

        // Contact button click listener
        contact.setOnClickListener(new View.OnClickListener() {
            @Override
            public void onClick(View view) {
            Intent intent = new Intent(Intent.ACTION_SENDTO);
            intent.setData(Uri.parse("mailto:xyz@gmail.com"));
            intent.putExtra(Intent.EXTRA_SUBJECT, "Subject");
            intent.putExtra(Intent.EXTRA_TEXT, "Email body");
            startActivity(intent);
            }
        });
    }
}
```
